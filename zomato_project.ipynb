# convert the data type of column _rate

```python
def zomatorate(value):
    value = str(value).split('/')
    value = value[0]  
    return float(value)  

```

```python
zomato['rate'] = zomato['rate'].apply(zomatorate)

zomato.head()
```

# 1) What type of restaurant do the majority of customers order from?

# Type of resturant

```python
zomato.head()
```

```python


# Count plot
sns.countplot(x=zomato['listed_in(type)'], palette=['red', 'yellow', 'purple', 'blue'])

# Labeling the x-axis
plt.xlabel('Type of Restaurant')
# most of the food is being ordered from dining restaurants. 


```

# 2 How many votes has each type of restaurant received from customers?


```python
zomato.head()
```

```python
zomato_data=pd.DataFrame(zomato.groupby('listed_in(type)')['votes'].sum())
```

```python
zomato_data=zomato_data.reset_index() #line chart
```

```python
zomato_data
```

```python
sns.relplot(zomato_data,x="listed_in(type)",y="votes",kind="line")
```

# 5) Which mode (online or offline) has received the maximum rating?


# which mode receives maximum rating

```python
zomato.head()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x='online_order',
           y='rate', data=zomato)
```

# conclusion - offline order received lower rating in  comparison to online order

# 6) Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?


```python
zomato.head()
```

```python
# Pivot table banana
pivot_table = zomato.pivot_table(index='listed_in(type)', 
                                  columns='online_order', 
                                  aggfunc='size', 
                                  fill_value=0)

# Heatmap banana
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Shubham-cafe')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```

# Conclusion: Dining restaurants primarily accept offline orders,whereas cafes primarily receive online orders. this suggests  that clients preferens order in person at restaurants,but prefer online ordering at cafes.
